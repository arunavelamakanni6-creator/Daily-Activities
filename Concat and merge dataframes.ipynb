## Concat DataFrames

```python
import pandas as pd
India_weather=pd.DataFrame({
    "City":["Mumbai","Bangalore","Vizag"],
    "Temperature":[30,29,39]
})
India_weather
```

```python
US_weather=pd.DataFrame({
    "City":["New york","Chicago","Orlando"],
    "Temperature":[32,45,35]
})
US_weather
```

```python
df=pd.concat([India_weather,US_weather])
df
```

```python
df=pd.concat([India_weather,US_weather],ignore_index=True)
df
```

```python
df=pd.concat([India_weather,US_weather],axis=1)
df
```

## Merge DataFrames

```python
temp_df=pd.DataFrame({
    "City":["Hyderabad","Vijayawada","Chennai","vizag"],
    "Temperature":[34,42,38,29]
})
temp_df
```

```python
humidity_df=pd.DataFrame({
    "City":["Hyderabad","Vijayawada","Chennai","Bangalore"],
    "Humidity":[62,65,60,75]
})
humidity_df
```

```python
df=pd.merge(temp_df,humidity_df,on="City")
df
```

```python
df=pd.merge(temp_df,humidity_df,on="City",how="outer")
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['Temperature']=df['Temperature'].fillna(df.Temperature.min())
df
```

```python
df.isnull()
```

```python
df['Humidity']=df['Humidity'].fillna(df.Humidity.max())
df
```

```python
df.isnull()
```
