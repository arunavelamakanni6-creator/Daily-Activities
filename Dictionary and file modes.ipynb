## FILE MODES

```python
f=open('abc.txt','w')
print('File name: ',f.name)
print('File mode: ',f.mode)
print(f.readable())
print(f.writable())
print(f.closed)
f.close()
print(f.closed)
```

## Read Mode

```python
with open("abc.txt", "r") as file:
    content=file.read()
print(content)
```

```python
with open("abc.txt", "r") as file:
    content=file.read(10)
print(content)
```

```python
with open("abc.txt", "r") as file:
    content=file.readline()
print(content)
```

```python
with open("abc.txt", "r") as file:
    content=file.readlines()
print(content)
```

## Write Mode

```python
with open("abc.txt", "w") as file:
    file.write("Aditya Degree College")
```

```python
l=["Aditya Degree College\n", "Kakinada\n", "Andhra Pradesh\n"]
with open("abc.txt","w") as file:
    file.writelines(l)
```

## Append Mode

```python
with open("abc.txt", "a") as file:
    file.write("533003")
```
